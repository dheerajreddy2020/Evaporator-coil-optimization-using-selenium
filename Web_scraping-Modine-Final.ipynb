```python
def get_inputs(i,nominal=False, cfm_variants=False,final_check=False):
    i=i-1
    if cfm_variants:
        design_conditions=pd.read_csv(loc+'inputs.csv')
        Tonnage=design_conditions['Tons'].iloc[i]
        Tdb_value=str(81)
        Twb_value=str(66)
        rows_value=str(int(design_conditions['Rows'].iloc[i]))
        circuit_value=str('-')+str(design_conditions['Circuits'].iloc[i])
        CFM_value=str(design_conditions['CFM'].iloc[i])
        if final_check:
            Tdb_value=str(81)
            Twb_value=str(71)
    elif nominal:
        design_conditions=pd.read_csv(loc+'inputs.csv')
        Tonnage=design_conditions['Tons'].iloc[i]
        Tdb_value=str(81)
        Twb_value=str(66)
        CFM_value=str(Tonnage*400)
        rows_value=str(int(design_conditions['Rows'].iloc[i]))
        circuit_value=str('-')+str(design_conditions['Circuits'].iloc[i])
    else:
        design_conditions=pd.read_csv(loc+'Design_conditions.csv')
        Tonnage=design_conditions['Tons'].iloc[i]
        Tdb_value=str(design_conditions['Tdb'].iloc[i])
        Twb_value=str(design_conditions['Twb'].iloc[i])
        rows_value=str(design_conditions['Rows'].iloc[i])
        circuit_value=str('-')+str(design_conditions['Circuits'].iloc[i])
        CFM_value=str(design_conditions['CFM'].iloc[i])

    Capacity_value=str(Tonnage*12)
    FH_numeric=design_conditions['FH'].iloc[i]
    if FH_numeric < 10:
        FH_value='0'+str(design_conditions['FH'].iloc[i])+'.00'
    else:
        FH_value=str(design_conditions['FH'].iloc[i])+'.00'
    FL_value=str(design_conditions['FL'].iloc[i])+'.00'
    OD_value='3/8'
    Tube_value=str(design_conditions['Tube_mat'].iloc[i])
    Fin_value=str(design_conditions['Fin_mat'].iloc[i])
    spacing_value=str(design_conditions['FPI'].iloc[i])
    surface_value=str(design_conditions['Surface'].iloc[i])
    #print(surface_value)
    return circuit_value,Tonnage,Capacity_value,FH_value,FL_value,OD_value,Tube_value,Fin_value,CFM_value,spacing_value,rows_value,surface_value,Tdb_value,Twb_value


def get_inputs_iteration(i):
    design_conditions=pd.read_csv(loc+'Design_conditions.csv')
    i=i-1
    Tonnage=design_conditions['Tons'].iloc[i]
    Capacity_value=str(Tonnage*12)
    FH_value=str(design_conditions['FH'].iloc[i])+'.00'
    FL_value=str(design_conditions['FL'].iloc[i])+'.00'
    Tube_value=str(design_conditions['Tube_mat'].iloc[i])
    Fin_value=str(design_conditions['Fin_mat'].iloc[i])
    CFM_value=str(design_conditions['CFM'].iloc[i])
    return Tonnage,Capacity_value,FH_value,FL_value,Tube_value,Fin_value,CFM_value
```

```python
daikin_limits=pd.read_csv(loc+'Daikin-limits.csv')

def convert_dtypes(df):
    df1=df.copy()
    df1['APD']=pd.to_numeric(df1['APD'])
    df1['Capacity']=pd.to_numeric(df1['Capacity'])
    df1['Face_velocity']=pd.to_numeric(df1['Face_velocity'])
    df1['FL']=pd.to_numeric(df1['FL'])
    df1['FH']=pd.to_numeric(df1['FH'])
    df1['Internal_volume']=df1['Internal_volume'].str.replace(',', '')
    df1['Ref_vel']=df1['Ref_vel'].str.replace(',', '')
    
    df1['Internal_volume']=pd.to_numeric(df1['Internal_volume'])
    df1['Ref_PD']=pd.to_numeric(df1['Ref_PD'])
    df1['Ref_vel']=pd.to_numeric(df1['Ref_vel'])
    df1['Circuits']=pd.to_numeric(df1['Circuiting'].str[2:4])
    df1['Rows']=pd.to_numeric(df1['Model_name'].str[5:7])
    df1['x_factor']=df1['FH']*df1['Rows']/df1['Circuits']
    df1.dropna(inplace=True)
    #df1.drop(['n_rows'],axis=1,inplace=True)
    return df1

def limits_check(df1,final=False):
    df=df1.copy()
    limits_df=daikin_limits[daikin_limits['Capacity (T)']==Tonnage]
    Min_vol=limits_df['Min-Vol'].iloc[0]
    Max_vol=limits_df['Max-Vol'].iloc[0]
    if final:
        Min_Cap=limits_df['Min-Capacity'].iloc[0]
        Max_Cap=limits_df['Max-Capacity'].iloc[0]
    else:
        Min_Cap=limits_df['Min-Capacity'].iloc[0]*0.90
        Max_Cap=limits_df['Max-Capacity'].iloc[0]*1.10
        df=df[df['APD']<1.0][df['Face_velocity']<530]
    df=df[df['Internal_volume']>Min_vol][df['Internal_volume']<Max_vol][df['Capacity']>Min_Cap][df['Capacity']<Max_Cap]
    #df=df[df['Capacity']>Min_Cap][df['Capacity']<Max_Cap]
    df=df[df['Ref_PD']<10.0][df['Ref_vel']>1000]
    #df=df[df['x_factor']%2==0]
    return df


def Capacity_range():
    limits_df=daikin_limits[daikin_limits['Capacity (T)']==Tonnage]
    Min_Cap=limits_df['Min-Capacity'].iloc[0]
    Max_Cap=limits_df['Max-Capacity'].iloc[0]
    return Min_Cap,Max_Cap

def Tonnage_check(capacity):
    for tons in daikin_limits['Capacity (T)']:
        max_cap=daikin_limits[daikin_limits['Capacity (T)']==tons]['Max-Capacity'].iloc[0]
        min_cap=daikin_limits[daikin_limits['Capacity (T)']==tons]['Min-Capacity'].iloc[0]
        if capacity < max_cap:
            if capacity > min_cap:
                out=tons
            else:
                out=None
            break
    return out
```
